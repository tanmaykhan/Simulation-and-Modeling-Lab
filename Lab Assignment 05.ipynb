Apply the One-Sample Kolmogorov-Smirnov (K-S) Test on a dataset of daily temperatures to
determine if the sample data comes from a specified normal distribution.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import kstest, norm
```

```python
# Step 1 & 2: Load the Dataset by Using the provided dataset in colab and Extract the temperatures from the dataset for analysis.
temperatures = np.array([15, 17, 16, 14, 18, 20, 19, 15, 21, 22])

# Step 3: Center and Scale the Data
mean_temp = np.mean(temperatures)
std_temp = np.std(temperatures)
scaled_temperatures = (temperatures - mean_temp) / std_temp

# Step 4: Perform the K-S Test
ks_statistic, p_value = kstest(scaled_temperatures, 'norm')
h = int(p_value < 0.05)

# Step 5: Output the Results
print(f"Test Decision (h): {h}")
print(f"K-S Statistic: {ks_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Step 6: Plot the Empirical CDF vs Standard Normal CDF
sorted_data = np.sort(scaled_temperatures)
empirical_cdf = np.arange(1, len(scaled_temperatures) + 1) / len(scaled_temperatures)

plt.figure(figsize=(8, 5))
plt.plot(sorted_data, empirical_cdf, marker='o', linestyle='-', label='Empirical CDF')
plt.plot(sorted_data, norm.cdf(sorted_data), color='red', linestyle='--', label='Standard Normal CDF')
plt.xlabel('Standardized Temperature')
plt.ylabel('Cumulative Probability')
plt.title('Empirical CDF vs Standard Normal CDF')
plt.legend()
plt.grid(True)
plt.show()

# Step 7: Conclusion
if h == 0:
    conclusion = "Fail to reject the null hypothesis: The data may come from a normal distribution."
else:
    conclusion = "Reject the null hypothesis: The data does not come from a normal distribution."

print("\nConclusion:")
print(conclusion)
```

The One-Sample Kolmogorov-Smirnov (K-S) test was applied to a 10-day temperature dataset after centering and scaling it to fit a standard normal distribution. The resulting p-value was greater than 0.05, indicating that there is no significant difference between the empirical distribution of the data and the theoretical standard normal distribution.

Therefore, we fail to reject the null hypothesis, suggesting that the sample data may have come from a normal distribution. However, it's important to note that "failing to reject" does not prove normality, but rather that there isn't strong enough evidence to say otherwise.
