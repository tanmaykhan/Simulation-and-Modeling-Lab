Data: [5.2, 5.4, 5.1, 5.6, 5.3, 5.8]

Hypothesized mean: 5.5

â€¢ Tasks :-
- Calculate the t-statistic and p-value.
- Interpret and visualize the result at a 0.05 significance level.

```python
import numpy as np
import scipy.stats as stats #new library
import matplotlib.pyplot as plt
```

```python
values = np.array([5.2, 5.4, 5.1, 5.6, 5.3, 5.8])
mean_value = 5.5

test_stat, p_val = stats.ttest_1samp(values, mean_value)

print(f"Sample Mean: {np.mean(values):.2f}")
print(f"Hypothesized Mean: {mean_value:.2f}")
print(f"t-Statistic: {test_stat:.2f}")
print(f"Degrees of Freedom: {len(values) - 1}")
print(f"p-Value: {p_val:.4f}")

alpha = 0.05
if p_val < alpha:
    print(f"The null hypothesis is rejected. The sample mean is significantly different from {mean_value:.2f}.")
else:
    print(f"The null hypothesis is not rejected. The sample mean is not significantly different from {mean_value:.2f}.")

plt.figure()
plt.bar(1, np.mean(values), color="lightblue", label='Sample Mean')
plt.errorbar(1, np.mean(values), yerr=np.std(values), fmt='k', linewidth=1.5)
plt.axhline(y=mean_value, color='r', linestyle='--', linewidth=1.5, label='Hypothesized Mean')
plt.title('Comparison of Sample Mean and Hypothesized Mean')
plt.xlim(0.5, 1.5)
plt.xlabel('Sample Value')
plt.ylabel('Mean Value')
plt.legend()
plt.grid(True)
plt.show()
```
